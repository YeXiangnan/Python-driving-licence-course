### 2.5 Image collection technique captures cracks in healable progress
At first，doing pre-cracking test to each concrete samples，then put them into different environments for curing.
After certain **exposure time(ET)**, collecting the geometries of cracks and calculate **Index of crack clousre(ICC)**

**Formula for evaluating the self-healing of cracks**

Crack width(CW):
$$CW=\frac{A_{crack}}{L_{crack,pre-crack}}$$

Index of crack clousre(ICC)：
$$ICC=1-\frac{A_{crack,healed}}{A_{crack,pre-crack}}$$

### 2.6 Calibration method for crack kinetic self-healing coefficient

Employing curve fitting techniques to analyse the Index of Crack Closure (ICC) of the crack area at various time intervals in the database：
$$ICC_{average}=1-e^{-S*t}$$

Where t indicates exposure time[month], S is self-healing coefficient for cracks under water environments which indicates the rate of crack closure.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Load data from Excel file
data = pd.read_excel(r"L:\My programs\Database for Crack closure\2023_09_08\Loading effect\Data_Load_CEM I + CA + (Nano).xlsx")

# Filter data for 'Salt water' exposure environment
tap_water1L = data[(data['Exposure environment'] == 'Tap water') & (data['Exposure time'] == 1) & (data['Load or No Load'] == 'Load')].copy()
tap_water1N = data[(data['Exposure environment'] == 'Tap water') & (data['Exposure time'] == 1) & (data['Load or No Load'] == 'No Load')].copy()
tap_water3L = data[(data['Exposure environment'] == 'Tap water') & (data['Exposure time'] == 3) & (data['Load or No Load'] == 'Load')].copy()
tap_water3N = data[(data['Exposure environment'] == 'Tap water') & (data['Exposure time'] == 3) & (data['Load or No Load'] == 'No Load')].copy()
tap_water6L = data[(data['Exposure environment'] == 'Tap water') & (data['Exposure time'] == 6) & (data['Load or No Load'] == 'Load')].copy()
tap_water6N = data[(data['Exposure environment'] == 'Tap water') & (data['Exposure time'] == 6) & (data['Load or No Load'] == 'No Load')].copy()


# Define crack width intervals
bins = [0, 20, 50, 100, 300]
tap_water1L.loc[:, 'Crack width(μm)'] = pd.cut(tap_water1L['Crack width(μm)'], bins=bins)
tap_water1N.loc[:, 'Crack width(μm)'] = pd.cut(tap_water1N['Crack width(μm)'], bins=bins)
tap_water3L.loc[:, 'Crack width(μm)'] = pd.cut(tap_water3L['Crack width(μm)'], bins=bins)
tap_water3N.loc[:, 'Crack width(μm)'] = pd.cut(tap_water3N['Crack width(μm)'], bins=bins)
tap_water6L.loc[:, 'Crack width(μm)'] = pd.cut(tap_water6L['Crack width(μm)'], bins=bins)
tap_water6N.loc[:, 'Crack width(μm)'] = pd.cut(tap_water6N['Crack width(μm)'], bins=bins)


# Store the means for each boxplot
boxplot_means_1L = []
boxplot_means_1N = []
boxplot_means_3L = []
boxplot_means_3N = []
boxplot_means_6L = []
boxplot_means_6N = []


# Boxplot
fig, ax = plt.subplots(figsize=(32, 12), dpi=400)

# Function to calculate and add bars to boxplots
def add_boxplot_bars(ax, box_data, positions, color, width):
    for i, interval in enumerate(box_data['Crack width(μm)'].unique()):
        box_values = box_data[box_data['Crack width(μm)'] == interval]['Crack closure']
        mean_value = box_values.mean()
        ax.boxplot(
            box_values,
            positions=[positions[i]],  # Adjust positions for each box
            showfliers=True,
            showmeans=True,
            widths=width,
            patch_artist=False,
            vert=True,
        )
        # Calculate the mean and store it
        boxplot_means = [mean_value]
        ax.bar(positions[i], boxplot_means[0], width=0.8, color=color, edgecolor='black', linewidth=1.2)
    
# Define positions for each box
positions_1L = [1,10,19,28]
positions_1N = [2,11,20,29]
positions_3L = [4,13,22,31]
positions_3N = [5,14,23,32]
positions_6L = [7,16,25,34]
positions_6N = [8,17,26,35]

# Add boxplots and bars for each dataset
add_boxplot_bars(ax, tap_water1L, positions_1L, 'blue', 0.4)
add_boxplot_bars(ax, tap_water1N, positions_1N, 'lightskyblue', 0.4)
add_boxplot_bars(ax, tap_water3L, positions_3L, 'green', 0.4)
add_boxplot_bars(ax, tap_water3N, positions_3N, 'lightgreen', 0.4)
add_boxplot_bars(ax, tap_water6L, positions_6L, 'red', 0.4)
add_boxplot_bars(ax, tap_water6N, positions_6N, 'pink', 0.4)


# Set x-axis ticks and labels
ax.set_xticks([4.5,13.5,22.5,31.5])
ax.set_xticklabels(['0-20', '20-50', '50-100', '100-300'], fontsize=30)
ax.set_xlabel('Crack Width [μm]', fontsize=30, weight='bold')
ax.set_ylabel('Index of Crack Closure', fontsize=30, weight='bold')
ax.set_ylim(0, 1.1)
plt.yticks(fontsize=24)

# Add vertical dashed lines
for i in range(0, 5):
    ax.axvline(i * 9, linestyle='-', color='black', alpha=1, linewidth=3.0)
for i in range(1, 12):
    ax.axvline(i * 3, linestyle='--', color='grey', alpha=.8)

# Remove grid lines
ax.grid(False)

# Set legend
legend_labels = ['1month+Load', '1month+No Load',  '3months+Load','3months+No load', '6months+Load', '6months+No load']
legend_colors = [ 'blue', 'lightskyblue', 'green','lightgreen', 'red', 'pink']

legend_patches = [Patch(color=color, label=label) for color, label in zip(legend_colors, legend_labels)]
ax.legend(handles=legend_patches, loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3, fontsize=28)

plt.show()
```

# 5 Calibration for self-healing kinetic law
Providing an example involving the following condtions:

Mix design: CEM I + CA;

Exposure environment: Geothermal water;

Width range: 50-100μm

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# data input
x_data = np.array([0, 1, 2, 3, 4, 5, 6])
y_data = np.array([0,0.364932245,0.437494193,0.608875027,0.765355179,0.804015801,0.885983821])
error_min = np.array([0,0.254112136,0.125510978,0.182162128,0.172996672,0.091284204,0.154737444])
error_max = np.array([0,0.327619747,0.157504952,0.220112676,0.172726267,0.134065645,0.097241811])

# Define the exponential fitting function
def exponential_fit(x, A):
    return 1 - np.exp(-A * x)

# Fitting the data
fit_params, _ = curve_fit(exponential_fit, x_data, y_data)

# Calculate the data points on the fitted curve
x_fit = np.linspace(0, 6, 100)
y_fit = exponential_fit(x_fit, *fit_params)

# Plotting data points and error bars
plt.errorbar(x_data, y_data, yerr=[error_min, error_max], fmt='o', capsize=5, label='Data')

# Plotting the fitted curve
plt.plot(x_fit, y_fit, 'r-', label='Fit')

# Display equations and R-squared values
plt.title('CEM I + CA  of Crack width 50-100μm exposed to Geothermal water')
plt.xlabel('Time [month]')
plt.ylabel('Crack Closure')
r_squared = 1 - np.sum((y_data - exponential_fit(x_data, *fit_params)) ** 2) / np.sum((y_data - np.mean(y_data)) ** 2)
equation_text = 'y = 1 - exp(-{:.2f}x)\nR-squared = {:.2f}'.format(fit_params[0], r_squared)
plt.text(3, 0.2, equation_text)

plt.legend()
plt.show()
```
